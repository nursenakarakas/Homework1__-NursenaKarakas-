```python
#Question 4
fmarno = resp['fmarno']

from collections import Counter
counter = Counter(fmarno)
print counter
```

```python
#or
import thinkstats2
hist=thinkstats2.Hist(fmarno)
for value in sorted(hist.Values()):
    print(value, hist[value])
import thinkplot
thinkplot.Hist(hist)
thinkplot.Show(xlabel='value', ylabel='frequency')
```

```python

#Question 5
nevermarried=resp[fmarno==0]
married=resp[fmarno>=1]

nevermarried_hist=thinkstats2.Hist(nevermarried.totincr)
married_hist=thinkstats2.Hist(married.totincr)

width=0.5
thinkplot.PrePlot(2)
thinkplot.Hist(nevermarried_hist, align='right', width=width)
thinkplot.Hist(married_hist, align='left', width=width)
thinkplot.Show(xlabel='Total Income', ylabel='Frequency')
```

```python
#Question 6
#for never married
max_total_income_nmarried=max(nevermarried.totincr)
min_total_income_nmarried=min(nevermarried.totincr)
nmarried_std=nevermarried.totincr.std()
nmarried_var=nevermarried.totincr.var()
nmarried_mean=nevermarried.totincr.mean()
print 'Never married pregnants have %d max , %d min , %d standard deviation, %d variance and %d mean values for total income.' %(max_total_income_nmarried,min_total_income_nmarried, nmarried_std,nmarried_var, nmarried_mean)
#for married
max_total_income_married=max(married.totincr)
min_total_income_married=min(married.totincr)
married_std=married.totincr.std()
married_var=married.totincr.var()
married_mean=married.totincr.mean()
print 'Married pregnants have %d max , %d min , %d standard deviation, %d variance and %d mean values for total income.' %(max_total_income_married,min_total_income_married, married_std,married_var, married_mean)
```
